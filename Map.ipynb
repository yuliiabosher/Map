```python
import geopandas as gpd
import pyogrio
import folium
```

```python
def read_shape_file(file):
  district = gpd.read_file(file, engine="pyogrio", columns=['name','type', 'geometry'])
  return district
```

```python
central_district = read_shape_file("https://download.geofabrik.de/russia/central-fed-district-latest-free.shp.zip")
display(central_district)
```

```python
central_district_military_buildings = central_district[central_district['type'] == 'military']
central_district_street_map = folium.Map(location = [55.755863, 37.617700], zoom_start=6)
folium.GeoJson(central_district_military_buildings.geometry).add_to(central_district_street_map)
display(central_district_street_map)
```

```python
central_district_military_buildings['center'] = [x.geometry.centroid for i, x in central_district_military_buildings.iterrows()]
display(central_district_military_buildings)
```

```python
central_district_military_buildings['location'] = [(row.center.y, row.center.x) for index, row in central_district_military_buildings.iterrows()]
central_district_military_buildings
```

```python
for index, row in central_district_military_buildings.iterrows():
  location = [row['location'][0], row['location'][1]]
  popup = f"<strong>{row['name']}</strong>"
  marker = folium.Marker(location=location, popup=popup)
  marker.add_to(central_district_street_map)
display(central_district_street_map)
```
