```python
import geopandas as gpd
import folium
import pyogrio
import pyarrow
```

```python
central_district = gpd.read_file("https://download.geofabrik.de/russia/central-fed-district-latest-free.shp.zip", \
                                              engine="pyogrio", use_arrow=True, columns=['name','type', 'geometry'])
display(central_district)
```

```python
central_district_military_buildings = central_district[central_district['type'] == 'military']
central_district_street_map = folium.Map(location = [55.755863, 37.617700], zoom_start=6)
folium.GeoJson(central_district_military_buildings.geometry).add_to(central_district_street_map)
display(central_district_street_map)
```

```python
central_district_military_buildings['center'] = [x.geometry.centroid for i, x in central_district_military_buildings.iterrows()]
display(central_district_military_buildings)
```

```python
kaliningrad_district = gpd.read_file('https://download.geofabrik.de/russia/kaliningrad-latest-free.shp.zip')
```

```python
print(kaliningrad_district)
```

```python
building_types_kaliningrad = kaliningrad_district['type'].unique()
display(building_types_kaliningrad)
kaliningrad_ships = kaliningrad_district[kaliningrad_district['type']=='ship']
display(kaliningrad_ships)
kaliningrad_transformer_tower = kaliningrad_district[kaliningrad_district['type']=='transformer_tower']
display(kaliningrad_transformer_tower)
kaliningrad_gasometers = kaliningrad_district[kaliningrad_district['type']=='gasometer']
display(kaliningrad_gasometers)
kaliningrad_manufacture = kaliningrad_district[kaliningrad_district['type']=='manufacture']
display(kaliningrad_manufacture)
kaliningrad_tower = kaliningrad_district[kaliningrad_district['type']=='tower']
display(kaliningrad_tower)
kaliningrad_industrial = kaliningrad_district[kaliningrad_district['type']=='industrial']
display(kaliningrad_industrial)
```

```python
kaliningrad_street_map = folium.Map(location = [54.715424, 20.509207], zoom_start=11)
folium.GeoJson(kaliningrad_industrial.geometry).add_to(kaliningrad_street_map)
display(kaliningrad_street_map)
```
