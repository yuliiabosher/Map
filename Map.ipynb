```python
import geopandas as gpd
import dask_geopandas
```

```python
kaliningrad_district_df = gpd.read_file('https://download.geofabrik.de/russia/kaliningrad-latest-free.shp.zip')
kaliningrad_district = dask_geopandas.from_geopandas(kaliningrad_district_df, npartitions=4)
```

```python
print(kaliningrad_district)
```

```python
building_types_kaliningrad = kaliningrad_district['type'].unique()
display(building_types_kaliningrad)
kaliningrad_ships = kaliningrad_district[kaliningrad_district['type']=='ship']
display(kaliningrad_ships)
kaliningrad_transformer_tower = kaliningrad_district[kaliningrad_district['type']=='transformer_tower']
display(kaliningrad_transformer_tower)
kaliningrad_gasometers = kaliningrad_district[kaliningrad_district['type']=='gasometer']
display(kaliningrad_gasometers)
kaliningrad_manufacture = kaliningrad_district[kaliningrad_district['type']=='manufacture']
display(kaliningrad_manufacture)
kaliningrad_tower = kaliningrad_district[kaliningrad_district['type']=='tower']
display(kaliningrad_tower)
kaliningrad_industrial = kaliningrad_district[kaliningrad_district['type']=='industrial']
display(kaliningrad_industrial)
```

```python
kaliningrad_district.geometry.area.compute()
kaliningrad_district = kaliningrad_district.spatial_shuffle()
kaliningrad_district
```
